```python
%load_ext autoreload
%autoreload 2
%env CUDA_VISIBLE_DEVICES=0
%matplotlib inline

import pickle

import torch

import nodegam
from nodegam.gams.vis_utils import vis_main_effects

device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
all_dfs = {}
for model in ['ga2m', 'gam', 'ebm', 'spline']:
    if model not in all_dfs:
        all_dfs[model] = []
    for fold in range(5):
        with open(f'resources/best_dfs/bikeshare_{model}_f{fold}.pkl', 'rb') as fp:
            all_dfs[model].append(pickle.load(fp))

    all_dfs[model] = nodegam.utils.average_GAM_dfs(all_dfs[model])
```

All main effects

```python
fig, ax = vis_main_effects(all_dfs)
```

Top 8 interactions

```python
fig, ax = vis_main_effects({'NODE-GA2M': all_dfs['ga2m']}, top_interactions=8, only_interactions=True)
```
